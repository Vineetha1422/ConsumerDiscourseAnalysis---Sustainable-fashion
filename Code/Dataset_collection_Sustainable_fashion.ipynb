```python
import praw
import pandas as pd
from datetime import datetime
import spacy
import time

import warnings
warnings.filterwarnings('ignore')
```

```python
# Initializing Reddit instance
reddit = praw.Reddit(client_id='{client_id}',
                     client_secret='{secret_key}',
                     user_agent='EGS vs Steam')

#Check
subreddit = reddit.subreddit('gaming')
for submission in subreddit.hot(limit=5):
    print(submission.title)
```

```python
subreddit_name = "SustainableFashion"

# Lists to store data
data = []
comments_collected = 0

subreddit = reddit.subreddit(subreddit_name)

# Function to count replies for a comment
def count_replies(comment):
    count = 0
    if hasattr(comment, 'replies'):
        for reply in comment.replies:
            count += 1
            count += count_replies(reply)  # Recursively count replies to replies
    return count

# Search for top posts
for query in ["sustainable fashion", "eco-friendly fashion", "green fashion"]:
    for sort_type in ["top", "controversial"]:
        for time_filter in ["all", "year"]:
            posts = subreddit.search(query, sort=sort_type, time_filter=time_filter, limit=100)

            for post in posts:
                post_title = post.title

                # Collect comments
                post.comments.replace_more(limit=None)
                for comment in post.comments.list():
                    if comments_collected < 1300:
                        reply_count = count_replies(comment)
                        data.append({
                            'post_title': post_title,
                            'comment_body': comment.body,
                            'comment_score': comment.score,
                            'comment_created_utc': comment.created_utc,
                            'comment_author': str(comment.author),
                            'reply_count': reply_count
                        })
                        comments_collected += 1
                    else:
                        break

                if comments_collected >= 1300:
                    break

            if comments_collected >= 1300:
                break
        if comments_collected >= 1300:
            break
    if comments_collected >= 1300:
        break

    # Sleep to avoid hitting rate limits
    time.sleep(2)

# Convert to DataFrame for easier manipulation and analysis
df = pd.DataFrame(data)

# Save to CSV for further analysis
df.to_csv('sustainable_fashion_data.csv', index=False)

print(f"Total comments collected: {comments_collected}")
```
